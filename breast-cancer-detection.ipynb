## Data visualisation

```python
# Created a pie chart to show malignant and benign %

x = cancer_data['diagnosis'].groupby(cancer_data['diagnosis']).count().index
cancer = cancer_data['diagnosis'].groupby(cancer_data['diagnosis']).count().values
colors = ["#d62728", "#a4d321"]
plt.pie(cancer, labels=x,  colors=colors , autopct='%1.1f%%', startangle=140)
plt.title("% of malignant and benign")
plt.show()
```

```python
# Selected the numerical variables of interest
numerical_vars = ['radius_mean', 'texture_mean', 'smoothness_mean']

# Filter the dataset for malignant diagnoses only
malignant_data = cancer_data[cancer_data['diagnosis'] == 'M']
```

```python
# Create histograms for each numerical variable
for col in numerical_vars:
    plt.figure(figsize=(8, 6))
    plt.hist(malignant_data[col], bins=10, color='pink', edgecolor='black')
        
    plt.xlabel(col.capitalize())
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col.capitalize()}')
    plt.show()
```

```python
# Create scatter plots to examine relationships between numerical variables
for i, var1 in enumerate(numerical_vars):
    for j, var2 in enumerate(numerical_vars):
        if i < j:
            plt.figure(figsize=(8, 6))
            plt.scatter(malignant_data[var1], malignant_data[var2], color='skyblue', edgecolor='black')
            plt.xlabel(var1)
            plt.ylabel(var2)
            plt.title(f'{var1} vs {var2}')
            plt.grid(True)
            plt.show()
```
